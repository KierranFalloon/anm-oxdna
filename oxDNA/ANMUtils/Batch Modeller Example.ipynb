## Prep Sequence Files we are aligning PDB info too

    As shown in the other Modeller tutorial we go to Unitprot and download the sequence files in fasta format
    for ALL chains in the PDB file
    
    Each Sequence file must renamed using the following convention 'chainX.fasta' where X is the capital
    letter of the Chain in the original PDB file
    
    ex. RNAP has chains G, H, I, J, K
        ->the corresponding sequence files will have the names:
            chainG.fasta
            chainH.fasta
            chainI.fasta
            chainJ.fasta
            chainK.fasta
            
    RECOMMENDED: change the title of the sequences in each fasta file (I normally choose the unitprot id 
    followed by the single letter chain ID)
            
    These files MUST be in the directory you call the function from

```python
#Here we call our first function
# This call aligns each PDB sequence to the sequences in the 'chainX.fasta' files and outputs them
# into Clustal alignment format into the same directory
modeltasks = mod.auto_prep_files('RNAP.pdb')
```

## We will use the Alignment files we just generated to tell Modeller which Residues we want to add

### We can edit the Alignment files and remove residues we do not want OR do not have enough restraints on their position (Usually Long chains of Residues on Beginning and Ends of the chain that are unobserved)

    Be Careful doing this, The code used here assumes every coordinate from the PDB file is used so you 
    cannot remove those residues
    
## Now that the Alignment Files have been edited to only solve for the residues we want we can call the next function

```python
# This function first generates the PIR file that Modeller uses in its automodeller 
# Next it generates models for all chains and outputs them into directory 'modX'
mod.auto_models(modeltasks, 'RNAP.pdb', model_number=5)
```

### Looking at the Output we can see how each of the 5 models for each chain scores, The best scoring models have Lower molpdf scores

## Unfortunately Modeller Outputs do not preserve B-factors or Chain IDs and actually saves into pdb models rather than pdb chains

### To combine the best scoring models back into a single file PDB, each chain must be in the original order of the original PDB file (for proper B-factor comparison)

### Using Chimera, open the best model for the first chain, then open the best models for the other chains (usually in Alphabetical Order) using File open

    Highly recommend renaming the best models and copying them to separate directory (Modeller's naming
    conventions are not great)
